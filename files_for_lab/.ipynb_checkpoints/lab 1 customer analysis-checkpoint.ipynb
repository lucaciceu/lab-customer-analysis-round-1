```python
#making sure all coluns are in lower case
df1.columns = df1.columns.str.lower()
df2.columns = df2.columns.str.lower()
df3.columns = df3.columns.str.lower()

#replacing 'ST' with 'state' for df1, df2
df1 = df1.rename(columns={'st':'state'})
df2 = df2.rename(columns={'st':'state'})

#verifying for df1
df1.columns
```

```python
# rearranging columns for df2 & df3 to have the same header as df1
df2 = df2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
df3 = df3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

#checking for df3
df3.columns
```

```python
#concatenate the files
df_total=pd.concat([df1, df2, df3])

#verifying
df_total.shape
```

```python
#showing data frame column types
df_total.dtypes
```

```python
#listing categorical columns
list_categorical_types = list(df_total.select_dtypes(include=['object']).columns)
list_categorical_types
```

```python
#listing numerical columns
list_numeric_types = list(df_total.select_dtypes(include=['float64']).columns)
list_numeric_types
```

```python
# deleting columns 'number of open complaints', 'education'
df_total = df_total.drop(['number of open complaints', 'education'], 1)
df_total.columns
```

```python
df_total['customer lifetime value']
```

```python
#checking for null values
df_total[df_total['customer lifetime value'].isna()==True]
```

```python
#checking the percentage of null values
df_total['customer lifetime value'].isna().sum()/len(df_total)*100
```

```python
#converting to string
df_total['customer lifetime value'].to_string()
#removing %
df_total['customer lifetime value'] = df_total['customer lifetime value'].str.replace('%','')
#converting null fields to zero
df_total['customer lifetime value'] = df_total['customer lifetime value'].fillna(0)
df_total['customer lifetime value']
```

```python
#cloning the df 
df_total_v2 = df_total.copy()
#converting the cloned df to numeric
df_total_v2['customer lifetime value']=pd.to_numeric(df_total_v2['customer lifetime value'], errors='coerce')
df_total_v2['customer lifetime value']
```

```python
#multiplying the values in the column by 100
df_total_v2['customer lifetime value']=100*df_total_v2['customer lifetime value']
df_total_v2
```

```python
#checking for duplicates (sum)
df_total_v2.duplicated().sum()
```

```python
#checking the duplicates
df_total_v2[df_total_v2.duplicated() == True]
```

```python
#droping duplicates
df_total_dropped_duplicates = df_total_v2.drop_duplicates()
df_total_dropped_duplicates
```

```python
#filtered by income 0 or negative
filtered_less_0 = df_total_dropped_duplicates[(df_total_dropped_duplicates['income']<=0)]
filtered_less_0
```
